# Data exploration and assessment - federation data

## Exploring the federation data set

```python
url = "https://www.powerliftingwatch.com/compare-federations"
```

```python
import pandas as pd
import numpy as np
```

```python
# Use pandas read_html to scrape data from html tables
dfs = pd.read_html(url)
```

```python
type(dfs)
```

```python
# returns a list of dfs one for each html table

dfs[0][0:10]
```

```python
# make a columns array from the header contained in the first 2 rows
df_cols = dfs[0].iloc[0:2,:].transpose()

# create an underscore separated, lower cased column name array

columns = (df_cols[0].replace(' ','_',regex=True) + df_cols[1].apply(lambda x: '' if isinstance(x, float) else '_' + str(x))).str.lower()
```

```python

df = pd.DataFrame(pd.concat([dfs[n][2:] for n in range(len(dfs))]))
df.columns = columns
df.reset_index()

```
